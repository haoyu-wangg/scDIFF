```python
import dataset
import torch
from torch.utils.data import DataLoader, Subset
```

```python
dataset
```

```python
# Define worker initialization function
def worker_init_fn(worker_id):
    worker_info = torch.utils.data.get_worker_info()
    dataset = worker_info.dataset  # Access the dataset object
    dataset._initialize_bwls(dataset.bw_path, dataset.bw_list)  # Initialize BigWig files in the current worker
```

```python

ds = dataset.SingleCellDataset(dataset.load_adata("/home/daozhang/ChIP-SANGO/preprocessed_data/LargeIntestineB_LargeIntestineA.h5ad"), seq_len=1344, genome="./genome/mm9.fa.h5", bw_path="/home/daozhang/ChIP-SANGO/data/LargeIntestine/", bw_list=["H3k4me1_mm9.bigWig", "H3k27ac_mm9.bigWig","H3k4me3_mm9.bigWig"])

# Create DataLoader
train_loader = DataLoader(
    ds,
    batch_size=128,
    shuffle=True,
    num_workers=12,  # High parallelism
    pin_memory=True,
    drop_last=True,
    prefetch_factor=4,
    worker_init_fn=worker_init_fn  # Initialize workers with BigWig files
)
```

```python
# Iterate through the DataLoader
for batch in train_loader:
    seq, epifeatures, target = batch
    print("Sequence shape:", seq.shape)
    print("Epifeatures shape:", epifeatures.shape)
    print("Target shape:", target.shape)
    break
```
